```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def Extract(filename):
    import re
    f = open(filename,'r').read()
    p = re.compile(ur'Iteration (\d+).*\n.*Test net output .* accuracy = (\d*\.\d*)\([\d|.]+\)\n')
    return np.array(re.findall(p,f)).astype(np.float)

def Plot(accuracy,label):
    line1 = plt.plot(accuracy[:,0],accuracy[:,1])
    plt.setp(line1,label=label)

Plot(Extract('./logs/2015-05-27_21_53_mit_indoor_nin_sc_finetuning.log'), 'given_tree')
Plot(Extract('./logs/2015-05-27_21_20_mit_indoor_nin_finetuning.log'), 'nin')

plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.title('Accuracy over Training Iteration - Mit_Indoor')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
axes = plt.gca()
#axes.set_ylim([0.55,0.68])
#axes.set_xlim([0,100000])
plt.show()
```
