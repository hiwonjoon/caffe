```python
def makeLabelKeyDict() :
    dic = {}
    for i in range(0,100):
        dic[i] = []
    
    import lmdb
    env = lmdb.open('/storage/datasets/mit-indoor/train_lmdb')
    with env.begin() as txn:
        cursor = txn.cursor()
        for key,value in cursor : 
            datum = caffe.proto.caffe_pb2.Datum()
            datum.ParseFromString(value)
            dic[datum.label].append(key)
    return dic

label_key = makeLabelKeyDict()
```

```python
def getImage(key):
    caffe_root = '../../'
    import lmdb
    
    imgs = [];
    
    img = np.zeros([256,256,3],np.uint8);
    env = lmdb.open('/storage/datasets/mit-indoor/train_lmdb')
    with env.begin() as txn:
        cursor = txn.cursor()
        value = cursor.get(key)
        datum = caffe.proto.caffe_pb2.Datum()
        datum.ParseFromString(value)
        
        label = datum.label
        temp = np.fromstring(datum.data,dtype=np.uint8).reshape(3,256,256)
        img[:,:,0] = temp[0,:,:];
        img[:,:,1] = temp[1,:,:];
        img[:,:,2] = temp[2,:,:];
        
    return (label,img)

def getActivation(keys):
    transformer = caffe.io.Transformer({'data': net.blobs['data'].data.shape})
    transformer.set_transpose('data', (2,0,1))
    transformer.set_mean('data', np.load('./imagenet_mean.npy').mean(1).mean(1)) # mean pixel
    transformer.set_raw_scale('data', 255)  # the reference model operates on images in [0,255] range instead of [0,1]
    transformer.set_channel_swap('data', (2,1,0))  # the reference model has channels in BGR order instead of RGB
    
    ret = {};
    ret['raw'] = {'loss' : np.zeros((len(keys),67)), 'pool4' : np.zeros((len(keys),67))}
    ret['avg'] = {'loss' : np.zeros(67), 'pool4' : np.zeros(67)}
    for i in range(0,len(keys),10):
        data = [];
        cnt = min(i+10,len(keys))-i;
        for j in range(i,i+cnt):
            label,raw = getImage(keys[j]);
            data.append(transformer.preprocess('data',raw))
        ret['raw']['loss'][i:i+cnt,:] = net.forward_all(data=np.asarray(data))['loss'].reshape(cnt,67)
        ret['raw']['pool4'][i:i+cnt,:] = net.blobs['pool4'].data[0:cnt,:,:,:].reshape(cnt,67)
    ret['avg']['pool4'] = np.sum(ret['raw']['pool4'],axis=0) / len(keys)
    return ret
```

```python
activations = np.zeros((67,67),np.float);
for i in range(0,67):
    activations[i,:] = getActivation(label_key[i][0:30])['avg']['pool4'];
```

```python
import sklearn.cluster as clu
NUM_CLUSTER = 25

from sklearn.neighbors import kneighbors_graph
#connectivity = kneighbors_graph(activations, n_neighbors=5)

#ward = clu.AgglomerativeClustering(n_clusters=NUM_CLUSTER,connectivity=connectivity).fit(activations)
ward = clu.AgglomerativeClustering(n_clusters=NUM_CLUSTER).fit(activations)
super_category = ward.labels_
for i in range(0,NUM_CLUSTER):
    for j in range(0,67):
        if super_category[j] == i:
            print labels[j], 
    print

for i in range(0,NUM_CLUSTER):
    print 'children {'
    for j in range(0,67):
        if super_category[j] == i:
            print '\tchildren {\n\t\tlabel:%d #%s\n\t}'%(j,labels[j])
    print '}'
```
